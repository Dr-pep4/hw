```python
def PP():
    N = int(input("숫자"))
    divisor_basket = []
    prime_basket = []
    count = 0
    
    ###########약수
    
    
    for i in range(1,N+1):
        if N%i == 0 :
            divisor_basket.append(i)
            
            
            
    ###########소수
    
    
    
    for a in range(2,N+1):
        for i in range(2, N+1):
            if (a)%i == 0:
                count += 1
        if count == 1:
            prime_basket.append(a)
        count = 0      
    print("약수 = ",divisor_basket)
    print("입력값 까지의 소수 = ",prime_basket)   
    
    ###########소인수 
    
    inter = list(set(divisor_basket)&set(prime_basket))
    
    print ("소인수 =",inter)
    
    
    ########### ugly
    
    minus = list(set(inter).difference([2,3,5]))
    if len(minus) == 0:
        print("ugly")
    else:
        print("not ugly")
    
    
    Prime_real = []
    index_num = []
    for i in inter:
        while N%i == 0:
            N = N/i
            Prime_real.append(i)
        index_num.append(Prime_real.count(i))
    
    
    for i in range(len(index_num)):
        print("(",inter[i],"^",index_num[i],")",end="")
    
PP()
```
